# 1. Probability

## 1.1. The Joint Proabbility Distributions

\begin{align*}
\mathrm {cov} [x, y] = \mathbb E [ (x - \mathbb E[x])(y - \mathbb E[y]) ]
\end{align*}

if $\mathbf x$ is a $d$-dimensional vector, then the result is:

\begin{align*}
\mathrm {cov}[ \mathbf x ] &= \mathbb E [ ( \mathbf x - \mathbb E[ \mathbf x  ])( \mathbf x - \mathbb E[ \mathbf x])^{\top} ] \\
      &= \begin{bmatrix}
    \mathrm{var}[x_1]       &  \mathrm{cov}[x_1, x_2] & \cdots & \mathrm{cov}[x_1, x_d] \\
    \mathrm{cov}[x_2, x_1]  &  \mathrm{var}[x_2]      & \cdots & \mathrm{cov}[x_2, x_d] \\
    \vdots                  &  \vdots                 & \ddots & \vdots \\
    \mathrm{cov}[x_d, x_1]  &  \mathrm{cov}[x_d, x_2] & \cdots & \mathrm{var}[x_d] 
\end{bmatrix}
\end{align*}

### The Multivariate Gaussian

In a $D$-dimensional space,

\begin{align*}
\mathcal N ( x|\mu, \Sigma ) = \frac {1} { (2 \pi)^{D/2} |\Sigma|^{1/2} } \exp \Big[ - \frac 1 2 (x - \mu)^{\top} \Sigma^{-1} (x - \mu)  \Big]
\end{align*}

## 1.2. Transformations of Random Variables

### 1.2.1. Linear Transformations

$ \mathbf y = f(\mathbf x) = \mathbf {Ax + b}  $

#### Mean 

\begin{align*}
\mathbb E[ \mathbf y] &= \mathbf A \mathbb E [ \mathbf x  ] + \mathbf b \\
                      &= \mathbf { A \mu + b  }                         
\end{align*}

if $y$ is scalar valued, 

\begin{align*}
\mathbb E[ \mathbf y] &= \mathbf a^{\top} \mathbb E [ \mathbf x  ] +  b \\
                      &= \mathbf { a^{\top} \mu  } + b                        
\end{align*}



#### Covariance

\begin{align*}
\mathrm {cov} [y] &= \mathrm {cov} [ \mathbf {Ax + b} ] \\
                  &= \mathbf { A \Sigma A^{\top} }
\end{align*}

\begin{align*}
\mathrm {cov} [y] &= \mathrm {cov} [ \mathbf {a^{\top}x } + b ] \\
                  &= \mathbf { a^{\top} \Sigma a }
\end{align*}

### 1.2.2. General Transformatioons

for a general transformation $y = f(x)$

#### Discrete Case

\begin{equation*}
p_y(y) = \sum_{x:f(x) = y} p_x(x)
\end{equation*}

#### Continuous Case

\begin{equation*}
p_y(y) = p_x(x)| \frac {dx} {dy} |
\end{equation*}


#### Multivariate Case

The Jacobian Matrix $\mathbf J_{ \mathbf x \rightarrow \mathbf y }$ is defined as:



\begin{equation*}
\mathbf J_{\mathbf x \rightarrow \mathbf y} = \frac {\partial(y_1, \ldots, y_n)} {\partial(x_1, \ldots, x_n)} = \begin{bmatrix}
    \frac {\partial y_1} {\partial x_1}   &  \ldots & \frac {\partial y_1} {\partial x_n} \\
    \vdots                                & \ddots & \vdots                               \\
    \frac {\partial y_n} {\partial x_1}   &  \ldots & \frac {\partial y_n} {\partial x_n} 
\end{bmatrix}
\end{equation*}

$$p_y(\mathbf  y) = p_x(\mathbf x) | \det \Big( \frac {\partial \mathbf x} {\partial \mathbf y} \Big) | = p_x(\mathbf x) | \det  \mathbf {J_{y \rightarrow x}} | $$

## 1.3. Information Theory

### 1.3.1. Entropy

$$ \mathbb H(x) = - \sum_{k=1}^{K}  {p(x=k) \log [p(x=k)}] $$

### 1.3.2. KL-Divergence

$$ \mathbb {KL} (p||q) = \sum_k p(k) \log \Big[ \frac {p(k)} {q(k)} \Big] $$

### 1.3.3. Mutual Information

$$ \mathbb I(X, Y) = \mathbb {KL} [ p(X,Y) || p(X)p(Y) ] = \sum_x \sum_y p(x, y) \log \Big[ \frac {p(x, y)} {p(x)p(y)} \Big] $$

## Eigendecomposition

\begin{equation*}
\mathbf {\Sigma = U A U^{\top}}
\end{equation*}

where $\mathbf U$ is an orthonormal matrix of eigenvectors, such that $\mathbf{U^{\top}U}=I$ and $\mathbf A$ is a diagonal matrix of eigenvalues. Thus, 

\begin{equation*}
\mathbf {\Sigma^{-1} = U^{-\top} A^{-1} U^{-1} = U A^{-1} U^{\top}} = \sum_{i=1}^{D} \frac {1} {\lambda_i} \mathbf u_i \mathbf u_i^{\top}
\end{equation*}

This also means that,

\begin{align*}
\mathbf { (x-\mu)^{\top} \Sigma^{-1} (x-\mu) } &= (x-\mu)^{\top} \Big( \sum_{i=1}^{D} \frac {1} {\lambda_i} \mathbf u_i \mathbf u_i^{\top} \Big) (x-\mu) \\
   &=  \sum_{i=1}^{D} \frac {1} {\lambda_i} (x-\mu)^{\top} \mathbf u_i \mathbf u_i^{\top}(x-\mu) \\
   &=  \sum_{i=1}^{D} \frac {y_i^2} {\lambda_i} 
\end{align*}
